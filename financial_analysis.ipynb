```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data
np.random.seed(0)
dates = pd.date_range(start='2023-01-01', periods=100)
values1 = np.random.rand(100)
values2 = np.random.randn(100)

# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'Value1': values1, 'Value2': values2})

# Visualization 1: Line plot
df.plot(x='Date', y='Value1', kind='line', color='blue', title='Line Plot')
plt.show()
```

```python
# Visualization 2: Bar plot
df.plot(x='Date', y='Value1', kind='bar', color='green', title='Bar Plot')
plt.show()
```

```python
# Visualization 3: Histogram
df['Value2'].plot(kind='hist', bins=20, color='orange', title='Histogram')
plt.show()
```

```python
# Visualization 4: Scatter plot
df.plot(x='Value1', y='Value2', kind='scatter', color='red', title='Scatter Plot')
plt.show()
```

```python
# Visualization 5: Box plot
df[['Value1', 'Value2']].plot(kind='box', color='purple', title='Box Plot')

plt.show()
```
